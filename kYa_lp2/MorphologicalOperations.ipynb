1. **What is the difference between an erosion and dilation??**  
- *Answer: An erosion eats away at the foreground object, while a dilation increases the size of the object*
- Erosion:  

- Dilation:  
  

Code example:

```python
img = cv.imread(r"./src/horseshoe_bend.png", 0)
img = cv.threshold(img, 0, 255, cv.THRESH_BINARY+cv.THRESH_OTSU)[1]
kernel = np.ones((3, 3))
img = cv.bitwise_not(img)

erosion = cv.erode(img, kernel, iterations=1)
kdm.show_rgb_image(erosion, "Erosion")

dilation = cv.dilate(img, kernel, iterations=1)
kdm.show_rgb_image(dilation, "Dilation")
```

2. **What is the difference between a closing and opening operation?**
- *Answer: An opening is an erosion followed by a dilation, while a closing is a dilation followed by an erosion*  
  
Code example:

```python
img = cv.imread(r"./src/horseshoe_bend.png", 0)

# Apply closing operation to the image
closing = cv.morphologyEx(img, cv.MORPH_CLOSE, kernel)

# Apply opening operation to the image
opening = cv.morphologyEx(img, cv.MORPH_OPEN, kernel)

# Show the original and processed images
kdm.show_rgb_image(img, "Original")
kdm.show_rgb_image(closing, "Closing")
kdm.show_rgb_image(opening, "Opening")
```

4. **How would you define a 5 x 20 rectangular kernel using OpenCV?**
- The correct answers are:
  - rectKernel = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 20))  
or
  - rectKernel = np.ones((5, 20), np.uint8)
  
First option uses the cv2.getStructuringElement function to create a rectangular kernel with the specified size. The second option uses the numpy.ones function to create a matrix of ones with the specified shape and data type. Both options will result in a 5 x 20 rectangular kernel that can be used for morphological operations.


```python
rectKernel = cv.getStructuringElement(cv.MORPH_RECT, (5, 20)) 
print(rectKernel)
```

5. **Use the OpenCV documentation to lookup the cv2.getStructuringElement function. Which kernel shape is NOT listed?**
- *Answer: MORPH_CIRCLE  

The cv2.getStructuringElement function can create three types of kernels: MORPH_RECT (rectangular), MORPH_ELLIPSE (elliptical) and MORPH_CROSS (cross-shaped). There is no option to create a circular kernel using this function.*
