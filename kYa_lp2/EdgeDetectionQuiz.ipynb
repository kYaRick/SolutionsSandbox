1. **The gradient magnitude representation of an image is very noisy and makes for a poor edge detector.**
- *Answer:  False*  

_Note_: The gradient magnitude representation of an image is actually a commonly used method for detecting edges. While it is true that the gradient magnitude representation can be noisy, there are techniques that can be used to reduce noise, such as smoothing the image before calculating the gradient. Additionally, the gradient magnitude provides useful information about the strength and location of edges in an image, which can be further processed and refined to detect edges more accurately. Therefore, the gradient magnitude representation is often a useful starting point for edge detection algorithms.


```python
# Load image in grayscale mode
img = cv.imread("./src/dog_contour.png", cv.IMREAD_GRAYSCALE)

# Apply Sobel operator to calculate gradients in x and y directions
grad_x = cv.Sobel(img, cv.CV_64F, 1, 0, ksize=3)
grad_y = cv.Sobel(img, cv.CV_64F, 0, 1, ksize=3)

# Calculate gradient magnitude
grad_mag = np.sqrt(grad_x**2 + grad_y**2)
grad_mag_norm = cv.normalize(grad_mag, np.zeros_like(grad_mag, dtype=np.float32), 0, 255, cv.NORM_MINMAX, dtype=cv.CV_16U)

kdm.show_rgb_image(img, "Input picture")
kdm.show_rgb_image(grad_mag_norm, "Edge detect")
```
