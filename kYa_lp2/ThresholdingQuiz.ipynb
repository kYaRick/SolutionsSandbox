_Note_: Adaptive thresholding can be faster than simple thresholding in some cases because it adjusts the threshold value for different regions of an image based on local characteristics, rather than using a fixed threshold value for the entire image.

2. **What is the primary limitation of Otsuâ€™s method for thresholding?**  
- *Answer: It assumes a bi-modal distribution of pixel intensities.*  

_Note_: The primary limitation of Otsu's method for thresholding is that it assumes a bi-modal distribution of pixel intensities in the input image. This means that the method may not work well for images with more complex distributions of pixel intensities or for images where the foreground and background are not well separated.  
Additionally, Otsu's method performs global thresholding, which means that a single threshold value is applied to the entire image. This can be limiting in cases where different regions of the image require different threshold values.


```python
# Load input image
img = cv.imread('./src/threshold_example.png', cv.IMREAD_GRAYSCALE)

kdm.show_rgb_image(img, 'Thresholded Image - origin')

ret, th = cv.threshold(img, 0, 255, cv.THRESH_BINARY)
kdm.show_rgb_image(th, 'Thresholded Image - without THRESH_OTSU')

ret, th = cv.threshold(img, 0, 255, cv.THRESH_BINARY + cv.THRESH_OTSU)
kdm.show_rgb_image(th, 'Thresholded Image - with THRESH_OTSU')
```


3. **What is a common nuisance of using simple thresholding via the cv2.threshold function?**
- *Answer:  A threshold value T must be manually supplied, making simple thresholding challenging in varying lighting conditions.*  

_Note_: A common nuisance of using simple thresholding via the cv2.threshold function is that a fixed threshold value T must be manually supplied, which can make simple thresholding challenging in varying lighting conditions or for images with a wide range of pixel intensities. This can lead to suboptimal results or even completely failed thresholding in some cases.


```python
ret,th = cv.threshold(img, 10, 255, cv.THRESH_BINARY)
kdm.show_rgb_image(th, 'Thresholded Image - th_k = 10')
print(f"Threshold value: {ret}")

ret,th = cv.threshold(img, 15, 255, cv.THRESH_BINARY)
kdm.show_rgb_image(th, 'Thresholded Image - th_k = 15')
print(f"Threshold value: {ret}")

ret,th = cv.threshold(img, 50, 255, cv.THRESH_BINARY)
kdm.show_rgb_image(th, 'Thresholded Image - th_k = 50')
print(f"Threshold value: {ret}")

ret,th = cv.threshold(img, 95, 255, cv.THRESH_BINARY)
kdm.show_rgb_image(th, 'Thresholded Image - th_k = 95')
print(f"Threshold value: {ret}")
```
